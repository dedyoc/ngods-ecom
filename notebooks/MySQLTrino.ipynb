```python
from trino.dbapi import connect
import trino
conn = connect(
    host='trino',  # The hostname should match the service name in docker-compose
    port=8060,     # The Trino coordinator port
    user='trino',
    catalog='mysql',
    schema='brazilian_ecommerce', 
)
# Create a cursor
cur = conn.cursor()

# Execute a SQL query
cur.execute('SELECT * FROM mysql.brazilian_ecommerce.sellers')

# Fetch the results
rows = cur.fetchall()

# Print the results
for row in rows:
    print(row)

# Close the cursor and connection
cur.close()
conn.close()
```
